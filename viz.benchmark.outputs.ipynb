```python
with open("results/benchmarks.output.txt") as f:
    lines = f.read().split("\n")
```

```python
!ls results/*benchmarks*
```

```python
benchmarks, arches, perfs = [], [], []

for line in lines:
    if "bench_linalg" in line:
        this_bench = convert_name(line)
    else:
        if "|" not in line or line.startswith("| arch") or line.startswith("|:-"):
            continue
        #print(this_bench, line)
        arch = line.split()[1]
        perf = float(line.split()[-2])

        benchmarks.append(this_bench)
        arches.append(arch)
        perfs.append(perf)
```

```python
def convert_name(line):
    if 'bench_linalg' not in line:
        return None
    elif line.startswith('bench_linalg'):
        return line[len('bench_linalg.'):]
    elif line.startswith('('):
        return line[len('bench_linalg.')+1:-1]
    else:
        raise ValueError
```

```python
import pandas as pd
df = pd.DataFrame({'benchmark': benchmarks, 'arch': arches, 'perf': perfs})
```
